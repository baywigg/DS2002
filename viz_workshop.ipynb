```python
import matplotlib.pyplot as mpl
import plotly.express as px
import pandas as pd
import csv
```

```python
data = []
with open("content/2023-2024 NBA Player Stats - Regular.csv", mode='r', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter=';')
    header = next(reader)
    try:
        for row in reader:
            data.append(row)
    except Exception as e:
        pass

df = pd.DataFrame(data, columns=header)
```

## Matplotlib (with pyplot)

```python
# Plot 1: Age distribution (Histogram)
mpl.figure(figsize=(10, 6))
mpl.hist(df['Age'].sort_values(), bins=8, color='orange', edgecolor='blue')
mpl.title("Player Age Distribution")
mpl.xlabel("Age")
mpl.ylabel("Frequency")
mpl.show()
```

```python
# Plot 2: Games played vs. Points scored (Scatter Plot)
df_sorted = df[['G', 'PTS']].sort_values(by='G')
mpl.figure(figsize=(10, 6))
mpl.scatter(df_sorted['G'], df_sorted['PTS'], color='green', alpha=0.7)
mpl.title("Games Played vs. Points Scored")
mpl.xlabel("Games Played")
mpl.ylabel("Points Scored")
mpl.xticks(ticks=range(0, 82, 5))  
mpl.yticks(ticks=range(0, 100, 10))
mpl.show()
```

## Plotly

```python
# Plot 1: Age vs Points Scored (Scatter Plot)
fig1 = px.scatter(df, x='Age', y='PTS', color='Pos', 
                  title="Age vs Points Scored by Position", 
                  labels={'Age': 'Player Age', 'PTS': 'Points Scored'},
                  color_discrete_sequence=px.colors.qualitative.Set1)
fig1.show()
```

```python
# Plot 2: Field Goals Made vs Free Throws Made (Scatter Plot)
fig2 = px.scatter(df, x='FG', y='FT', color='Tm', 
                  title="Field Goals Made vs Free Throws Made by Team", 
                  labels={'FG': 'Field Goals Made', 'FT': 'Free Throws Made'},
                  color_discrete_sequence=px.colors.qualitative.Set2)
fig2.show()
```
